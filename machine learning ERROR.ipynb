# numpy generating random numbers

```python
X_1 = np.random.randint(0, 30, size=60, dtype='l')
```

```python
y_1 = np.random.randint(0, 30, size=60, dtype='l')
```

```python
X_2 = np.random.randint(29, 60, size=60, dtype='l')
```

```python
y_2 = np.random.randint(29, 40, size=60, dtype='l')
```

```python
X_3 = np.random.randint(59, 90, size=60, dtype='l')
```

```python
y_3 = np.random.randint(39, 60, size=60, dtype='l')
```

```python
X = np.concatenate((X_1, X_2, X_3), axis=0)
```

```python
y = np.concatenate((y_1, y_2, y_3), axis=0)
```

```python
y_1_sort = y_1.argsort()
y_1_min_index = y_1_sort[0:7]
y_1_min = y_1[y_1_min_index]
X_1_min = X_1[y_1_min_index]

y_2_sort = y_2.argsort()
y_2_min_index = y_2_sort[0:7]
y_2_min = y_2[y_2_min_index]
X_2_min = X_2[y_2_min_index]

y_3_sort = y_3.argsort()
y_3_min_index = y_3_sort[0:7]
y_3_min = y_3[y_3_min_index]
X_3_min = X_3[y_3_min_index]

X_min = np.concatenate((X_1_min, X_2_min, X_3_min), axis=0)
y_min = np.concatenate((y_1_min, y_2_min, y_3_min), axis=0)
```

```python
plt.scatter(X, y)
plt.scatter(X_min, y_min)
```

```python
y_1_sort = y_1.argsort()[::-1]
y_1_max_index = y_1_sort[0:7]
y_1_max = y_1[y_1_max_index]
X_1_max = X_1[y_1_max_index]

y_2_sort = y_2.argsort()[::-1]
y_2_max_index = y_2_sort[0:7]
y_2_max = y_2[y_2_max_index]
X_2_max = X_2[y_2_max_index]

y_3_sort = y_3.argsort()[::-1]
y_3_max_index = y_3_sort[0:7]
y_3_max = y_3[y_3_max_index]
X_3_max = X_3[y_3_max_index]

X_max = np.concatenate((X_1_max, X_2_max, X_3_max), axis=0)
y_max = np.concatenate((y_1_max, y_2_max, y_3_max), axis=0)
```

```python
plt.scatter(X, y)
plt.scatter(X_min, y_min)
plt.scatter(X_max, y_max)
```

# Polynomial Reg main model

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg_1 = PolynomialFeatures(degree = 6)
X_poly = poly_reg_1.fit_transform(X.reshape(-1, 1))
```

```python
poly_reg_1_fit = poly_reg_1.fit(X_poly, y)
```

```python
lin_reg_1 = LinearRegression()
```

```python
lin_reg_1.fit(X_poly, y)
```

```python
y_pred_poly_1 = lin_reg_1.predict(X_poly)
```

```python
# sorting for plotting
X_sort = X.argsort()
X_plot = X[X_sort]
y_pred_poly_1_plot = y_pred_poly_1[X_sort]
```

```python
plt.figure(figsize = (15,8))
plt.scatter(X,y)
plt.plot(X_plot, y_pred_poly_1_plot, color = 'r')
plt.title('Polynomial Regression')
plt.xlabel('X')
plt.ylabel('y')
```

# Polynomial Reg model maximum boundary

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg_3 = PolynomialFeatures(degree = 3)
X_poly_max = poly_reg_3.fit_transform(X_max.reshape(-1, 1))
```

```python
poly_reg_3_fit = poly_reg_3.fit(X_poly_max, y_max)
```

```python
lin_reg_3 = LinearRegression()
```

```python
lin_reg_3.fit(X_poly_max, y_max)
```

```python
y_pred_poly_3 = lin_reg_3.predict(X_poly_max)
```

```python
# sorting for plotting
X_max_sort = X_max.argsort()
X_max_plot = X_max[X_max_sort]
y_pred_poly_3_plot = y_pred_poly_3[X_max_sort]
```

```python
plt.figure(figsize = (15,8))
plt.scatter(X, y)
plt.scatter(X_min, y_min)
plt.scatter(X_max, y_max)
plt.plot(X_plot, y_pred_poly_1_plot)
plt.plot(X_min_plot, y_pred_poly_2_plot)
plt.plot(X_max_plot, y_pred_poly_3_plot)
plt.title('Polynomial Regression with error margin')
plt.xlabel('X')
plt.ylabel('y')
```

```python
predicto = lin_reg_1.predict(poly_reg_1.fit_transform(20))
predicto_plus = lin_reg_3.predict(poly_reg_3.fit_transform(20)) - predicto
predicto_minus = predicto - lin_reg_2.predict(poly_reg_2.fit_transform(20))
```
